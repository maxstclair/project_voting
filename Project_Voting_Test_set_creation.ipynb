```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import linear_model

# Import Data
df = pd.read_csv('./project_voting/data/Vot_Census_Data_new.xlsx - Sheet1.csv')

# Dummies for fixed effects
dummies = pd.get_dummies(df['COUNTY'])
df2 = pd.concat([df, dummies], axis=1)
df2.head()
```

```python
from sklearn.linear_model import LinearRegression

# Project the 2024 values of the census metrics, essentially creating the "test" set (though we don't have actual outcomes)

stats = ["Total Population", "Income to Poverty Level Ratio", 'Per Capita Income vs Pop', 'Median House Value', 'GED Ratio']
df_2024 = pd.DataFrame(columns=["County","Total Population", "Income to Poverty Level Ratio", 'Per Capita Income vs Pop', 'Median House Value', 'GED Ratio'])
df_2024['County'] = df['COUNTY'].unique()

for stat in stats:
  counter = 0
  for county in df["COUNTY"]:
    df3 = df[df["COUNTY"] == county]
    X = df3["YEAR"].to_frame()
    y = df3[stat].to_frame()

    lin_reg = LinearRegression()
    lin_reg.fit(X,y)

    predicted = lin_reg.predict(pd.DataFrame({"YEAR":[2024]}))

    df_2024[stat][counter] = predicted[0][0]
    counter+=1
```
